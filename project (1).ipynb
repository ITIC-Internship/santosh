```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import re
import string
```

```python
def preprocess_data(text):
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = text.lower()
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('\n', '', text)
  
    
    return text
```

```python
df['review']=df['review'].apply(preprocess_data)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['sentiment']=le.fit_transform(df['sentiment'])
```

```python
df
```

```python
firstcol=df['review']
secondcol=df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
firstcol_train,firstcol_test,secondcol_train,secondcol_test=train_test_split(firstcol,secondcol,random_state=0,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
```

```python
firstcol_train_trans=cv.fit_transform(firstcol_train)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(firstcol_train_trans,secondcol_train)
pred_secondcol=lr.predict(cv.transform(firstcol_test))
```

```python
from sklearn.metrics import accuracy_score
score_1=accuracy_score(secondcol_test,pred_secondcol)
```
